```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
world = pd.read_csv('gapminder.csv', index_col =0)
```

```python
gdp_cap = world.gdp_cap
life_exp = world.life_exp
pop = world.population
```

```python
np_pop = np.array(pop)
np_pop = np_pop * 2 / 1000000
```

```python
continent = {'Asia' : 'red',
           'Europe' : 'blue',
           'Africa' : 'green',
           'Americas' : 'black',
           'Oceania' : 'yellow'}
```

```python
plt.scatter(gdp_cap, life_exp, s=np_pop, alpha = 0.7)
plt.xscale('log')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

plt.title('World Development 2007')
plt.xlabel('GDP per Capita [In USD]')
plt.ylabel('Life Expetancy [In Years]')

plt.show()
```
